```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Jupyter notebook
# dict = {"Jefke": [1,4,5,8,3], "Mo":[2,3,4,2,0], "Isabel": [9,2,6,3,1]}
# data =  ["Pooh", "Winnie", "Kanga", "Roo"]
# index = ["P1", "P2", "P3"]
# df = pd.DataFrame(dict)

data = {
    'year':['2020', '2021', '2022', '2023', '2024'],
    'pigeons':[33,52,26,40,38],
    'sparrows':[54,48,66,59,61],
    'others':[104,98,88,92,101]
}

```

```python
import pandas as pd
# stap 1: maak dataframe
df = pd.DataFrame(data)
df
```

```python
plt.figure()
plt.plot(df["year"], df["pigeons"], marker = "o", label = 'pigeon', color = 'blue')
plt.plot(df["year"], df["sparrows"],marker = ">", linestyle = 'dotted', color = 'green', markeredgecolor = 'red', label = 'sparrow')
plt.plot(df["year"], df["others"], marker= 'o', label = 'other', linestyle = 'dashed', color = 'red', fillstyle = 'top', markeredgecolor = 'grey')

plt.xlabel("jaar")
plt.ylabel("aantal")
plt.title("Aantal vogels gespot in 5 jaar")
plt.legend()


```

```python
plt.figure()
plt.bar(df['year'], df['pigeons'] + df['sparrows'] + df['others'], width = 0.4, color = 'purple', label = 'totaal')
plt.xlabel('jaar')
plt.ylabel('aantal')
plt.title('Aantal vogels per jaar')
plt.legend()
```

```python
plt.figure()

plt.bar(df['year'], df['pigeons'] + df['sparrows'] + df['others'], width = -0.4, color = 'purple', label = 'totaal', align = 'edge')
plt.bar(df['year'], df['pigeons'], width = 0.4, color = 'green', label = 'pigeons', align = 'edge')
plt.bar(df['year'], df['sparrows'], bottom = df['pigeons'], width = 0.4, color = 'blue', label = 'sparrows', align = 'edge')
plt.bar(df['year'], df['others'], bottom = df['pigeons']+df['sparrows'], width = 0.4, color = 'orange', label = 'others', align = 'edge')
plt.xlabel('jaar')
plt.ylabel('aantal')
plt.title('Aantal vogels per jaar')
plt.legend()
```
